```python
dfrom numpy import *
from numpy.fft import fft, ifft, fftshift, ifftshift
from numpy.lib import scimath as SM
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

```python
def dy_dt(y): #y'=-k^2y with k=2
    return -4*y #this is the right hand side of the heat equation after fourier
######
#tmax is the maximum value of t
#h is the steo size along t axis
#y0 is the initial data
######

def euler_path(tmax, h, y0):
    npts = int(tmax/h) #number of points
    path = zeros((npts+1), dtype=float)
    path[0] = y0 #the first entry of the approximated solution
    y=y0
    for i in range(npts):
        y = y + h*dy_dt(y)
        path[i+1] = y
    return path
################################################################
################################################################
###### Runga Kutta Definition ##################################
################################################################
################################################################


```

```python
y1 = euler_path(2, 0.25, 2)
t1 = linspace(0, 2, int(2/0.25)+1)

y2 = euler_path(2, 0.01, 2)
t2 = linspace(0, 2, int(2/0.01)+1)

t = linspace(0, 2, 100)
y3 = 2*exp(-4*t)


######
plt.plot(t, y3, 'r-', label='Exact')
plt.plot(t1, y1, 'b^-', label='h=0.25')
plt.plot(t2, y2, 'm*-', label='h=0.01')
plt.xlabel('t')
plt.ylabel('y')
plt.title('Euler Method for Heat')
plt.legend(loc=1)
plt.savefig('eulerexample.png')
```
